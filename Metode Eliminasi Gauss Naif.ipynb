```
n = 3
```

```
A = np.array([(4,-1,2,3),(0,-2,7,-4),(0,0,6,5),(0,0,0,3)])
```

```
B = np.array([(20),(-7),(4),(6)]).reshape(n+1,1)
```

```
B
```

```
x = np.zeros((n+1,1))
x
```

```
"Teknik Penyulihan mundur"
x[n] = B[n]/A[n,n]
for k in range(n-1,-1,-1):
  sigma = 0
  for j in range(k+1,n+1,1):
    sigma = sigma + A[k,j]*x[j]
  x[k] = (B[k] - sigma)/A[k,k]
```

```
def sulih_mundur(A,B,orde):
  n = orde-1
  x = np.zeros((n+1,1))
  "Teknik Penyulihan mundur"
  x[n] = B[n]/A[n,n]
  for k in range(n-1,-1,-1):
    sigma = 0
    for j in range(k+1,n+1,1):
      sigma = sigma + A[k,j]*x[j]
    x[k] = (B[k] - sigma)/A[k,k]
  return x
```

```
n = orde-1
```

```
sulih_mundur(A,B,4)
```

```
a = np.array([(2,3,-1),(4,4,-3),(-2,3,-1)])
orde = 3
a
```

```
b = np.array([(5),(3),(1)]).reshape(orde,1)
b
```

```
for k in range(0,n):
  for i in range(k+1,n+1):
    m = a[i,k]/a[k,k]
    for j in range(k,n+1):
      a[i,j] = a[i,j] - (m*a[k,j])
    b[i] = b[i]-(m*b[k])
sulih_mundur(a,b,orde)
```
