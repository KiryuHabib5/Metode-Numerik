```
fungsi = input("Masukan Fungsi :")
def f(x):
  e = np.e
  return eval(fungsi)

dfungsi = input("Masukan Turunan Pertama Fungsi")
def df(x):
  e = np.e
  return eval(dfungsi)

d2fungsi = input("Masukan Turunan Kedua Fungsi")
def d2f(x):
  e = np.e
  return eval(d2fungsi)
```

```
e = float(input("Masukan Toleransi maksimum : "))
N = int(input("Masukan Iterasi Maksimum : "))
```

```
f(1.10526315789473694728)
```

```
X_awal = float(input("Masukan Tebakan awal : "))
```

```
print("|X_awal\t\t\t| X_baru\t\t| |X_awal-X_baru|\t|")
print("========================================================================|")
for i in range(N):
  X_baru = X_awal - ((f(X_awal)*df(X_awal))/(df(X_awal)**2-(d2f(X_awal)*f(X_awal))))
  if f(X_baru) == 0:
    print("========================================================================|")
    print("Memungkinkan terdapat pembagian nol")
    print("Akar Taksiran : %0.20f" %(X_baru))
    break
  print("|%0.20f |%-0.20f |%-0.20f |" %(X_awal,X_baru,np.abs(X_awal-X_baru)))
  if np.abs(X_awal-X_baru) < e:
    print("========================================================================|")
    print(np.abs(X_awal-X_baru))
    print("Akar Taksiran : %0.20f" %(X_baru))
    break
  elif i > N-1:
    print("========================================================================|")
    print("Leleran Divergen")
    break
  X_awal = X_baru
print("========================================================================|")
  
```
