```
print("|Batas Bawah\t\t|Batas Atas \t\t|X\t\t\t| iterasi |")
print("==================================================================================|")
for i in range(1,N):
  if f(batas_bawah)-f(batas_atas) == 0:
    print("Terdapat pembagian dengan angka nol")
    break
  else:
    xr = (f(batas_bawah)*batas_atas - f(batas_atas)*batas_bawah)/(f(batas_bawah)-f(batas_atas))
    print("|%0.20f |%-0.20f |%-0.20f |\t%d|" %(batas_bawah,batas_atas,xr,i))
    if np.abs(f(xr)) <= e:
      break
    elif f(batas_bawah)*f(xr) <= 0:
      batas_bawah = xr
    else:
      batas_atas = xr
print("==================================================================================|")
```
