```
def inperpolasi_newton(x,y,orde,x_input):
  ST = np.zeros((orde,orde))
  n = orde - 1
  for i in range(orde):
    ST[i,0] = y[i]

  for k in range(1,orde):
    for i in range(0,orde-k):
      #ST[i,k] = (ST[i+1,k-1]-ST[i,k-i])/(x[i+k]-x[i])
      ST[i,k] = (ST[i+1,k-1]-ST[i,k-1])/(x[i+k]-x[i])
  print(ST)

  jumlah = ST[0,0]
  for i in range(1,n):
    suku = ST[0,i]
    for k in range(0,i):
      suku = suku*(x_input-x[k])
    jumlah = jumlah + suku
  newton = jumlah
  return newton
```
