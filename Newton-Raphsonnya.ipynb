```
import numpy as np
import math
```

```
fungsi = input("Masukan fungsi : ")
def f(x):
  e = np.e
  return eval(fungsi)

Dfungsi = input("Masukan Turunan dari fungsi : ")
def df(x):
  e = np.e
  return eval(Dfungsi)
```

```
e = float(input("Masukan Toleransi maksimum : "))
N = int(input("Masukan Iterasi maksimum : "))
```

```
X_lama = float(input("Masukan Tebakan awal : "))
```

```
print("|X_lama\t\t\t| X_baru\t\t| |X_lama-X_baru|\t|")
print("========================================================================|")
for i in range(N):
  if df(X_lama) == 0:
    print("Terdapat pembagian dengan angka nol")
    break
  X_baru = X_lama - (f(X_lama)/df(X_lama))
  print("|%0.20f |%-0.20f |%-0.20f |" %(X_lama,X_baru,np.abs(X_lama-X_baru)))
  if np.abs(X_lama-X_baru) < e:
    print("========================================================================|")
    print("Akar Taksiran : %0.20f" %(X_baru))
    break
  elif i > N-1:
    print("========================================================================|")
    print("Leleran Divergen")
    break
  X_lama = X_baru
print("========================================================================|")
```
