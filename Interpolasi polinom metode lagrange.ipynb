```
def lagrange_interpolasi(orde,x_input):
  n = orde - 1
  x = np.zeros(orde)
  y = np.zeros(orde)
  for i in range(orde):
    print("Masukan x[",i,"]",end="")
    x[i] = float(input(" : "))
    print("Masukan y[",i,"]",end="")
    y[i] = float(input(" : "))
  L = 0
  for i in range(orde):
    pi = 1
    for j in range(orde):
      if i != j:
        #print(pi ,"*", (x_input,"-",x[j]),"/",(x[i],"-",x[j]))
        pi = pi * (x_input-x[j])/(x[i]-x[j])
    #print(L ,"+", y[i],"*",pi)
    L = L + y[i]*pi
  Lagrange = L
  return Lagrange
```
