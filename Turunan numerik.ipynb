```
#formula turunan dengan selisih maju
def d_maju(f,x,h):
  return (f(x+h)-f(x))/h
```

```
#formula turunan dengan selisih mundur
def d_mundur(f,x,h):
  return (f(x)-f(x-h))/h
```

```
d_mundur(f,1,0.1)
```

```
#formula turunan selisih terpusat
def d_pusat(f,x,h):
  return (f(x+h)-f(x-h))/(2*h)
```

```
d_pusat(f,1,0.1)
```
