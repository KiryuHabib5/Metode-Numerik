```
#horner rules
def P(x):
  input_user = input("Masukan koef dari variabel paling tinggi (pisahkan dengan koma) : ")
  koef = np.array(input_user.split(','))

  n = len(koef)-1
  for i in range(n):
    if i==0:
      px = a[i]*x + a[i+1]
    else:
      px = px * x + a[i+1]

  for i in range(len(koef)):
    if i == 0:
      print("P(x) = (",koef[i],")x ^",n,end=" ")
      n-=1
    elif i<len(koef)-1:
      print("+ (",koef[i],")x ^",n,end=" ")
      n-=1
    else:
      print("+ (",koef[i],")",end=" ")
  print("\nP(",x,") = ",px)
  return px
```
